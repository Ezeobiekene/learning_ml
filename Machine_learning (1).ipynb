```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x, y):
    m_curr = b_curr = 0
    learning_rate = 0.0002
    iterations = 10000
    n = len(x)
    cost_history = []
    lines_to_plot = []  # store (m, b) at key steps

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        cost = (1/n) * sum((y - y_predicted) ** 2)
        md = -(2/n) * sum(x * (y - y_predicted))
        bd = -(2/n) * sum(y - y_predicted)
        m_curr -= learning_rate * md
        b_curr -= learning_rate * bd
        cost_history.append(cost)

        # Save current slope and intercept every 1000 iterations
        if i % 1000 == 0 or i == iterations - 1:
            lines_to_plot.append((m_curr, b_curr))

    # Plot the data points
    plt.figure(figsize=(8, 6))
    plt.scatter(x, y, color='red', label='Data points')

    # Plot lines from different stages
    x_vals = np.linspace(min(x), max(x), 100)
    for idx, (m, b) in enumerate(lines_to_plot):
        y_vals = m * x_vals + b
        label = f'Iter {idx * 1000}' if idx < len(lines_to_plot) - 1 else f'Final Line'
        plt.plot(x_vals, y_vals, label=label)

    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Gradient Descent Line Fit Over Time")
    plt.legend()
    plt.grid(True)
    plt.show()

x = np.array([1, 3, 5, 6, 9])
y = np.array([4, 6, 7, 11, 13])

gradient_descent(x, y)
```

# Video 20 - Bias vs Variance

* Overfit model tend to have high variance.
* High variance means the mode is sensitive to the changes in the training data.
  - AKA the test error differ greatly based on the selected training data

* Bias is the measurement of how accurately a model can capture a pattern in a training dataset. (train error)
  - When the training error is big it is said to be a high bias


* The goal is to have a low bias and low variance

* Overfit models make high variance while Underfit models make high bias
* Balanced fit models make low variance and low bias
* There is a trade off between bias and variance

image.png


Ways to get balanced fit model
1. Cross validation
2. Regularization
3. Dimentionality Reduction (PCA)
4. Ensemble Techniques

# Machine Learning Algorithms

| **Algorithm**                                   | **Definition**                                                                                                                                                                         |
| ----------------------------------------------- | -------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **Linear Regression**                           | A supervised regression algorithm that models the linear relationship between input features and a continuous target using the equation: **y = wX + b**.                               |
| **Logistic Regression**                         | A supervised classification algorithm that models the probability of class membership using the sigmoid function; useful for binary classification.                                    |
| **K-Nearest Neighbors (KNN)**                   | A non-parametric, instance-based learning algorithm that classifies a data point based on the **majority class of its k closest neighbors** using a distance metric (e.g., Euclidean). |
| **Naive Bayes**                                 | A probabilistic classifier based on **Bayes’ Theorem**, assuming strong (naive) independence between features. It works well for text data like spam filtering.                        |
| **Decision Tree**                               | A tree-based model that recursively splits data on feature values using impurity metrics like **Gini index** or **entropy** to maximize information gain.                              |
| **Random Forest**                               | An ensemble method that builds multiple decision trees using bootstrapped samples and random feature subsets, and aggregates their results (majority vote or average).                 |
| **Support Vector Machine (SVM)**                | A powerful classifier that finds the **optimal hyperplane** which maximizes the margin between classes; uses **kernel tricks** for non-linear data.                                    |
| **Gradient Boosting (e.g., XGBoost, LightGBM)** | An ensemble of decision trees built sequentially, where each new tree focuses on reducing the errors (residuals) of the previous one using gradient descent.                           |
| **K-Means Clustering**                          | An unsupervised algorithm that partitions data into **k clusters** by minimizing the **intra-cluster variance** using centroid updates.                                                |
| **PCA (Principal Component Analysis)**          | A linear dimensionality reduction technique that projects data into a lower-dimensional space by maximizing variance along new axes (principal components).                            |
| **Neural Networks (MLP)**                       | Composed of layers of interconnected neurons (units), neural nets learn non-linear relationships through weight updates using **backpropagation** and **gradient descent**.            |
| **CNN (Convolutional Neural Network)**          | A specialized neural network architecture for image data that uses **convolutions**, **pooling**, and **filters** to detect spatial hierarchies and patterns.                          |
| **RNN (Recurrent Neural Network)**              | A type of neural network designed for **sequential data**, where outputs depend on previous time steps; uses **hidden states** to capture temporal relationships.                      |
| **Transformers (e.g., BERT, GPT)**              | A deep learning architecture that uses **self-attention mechanisms** to capture long-range dependencies in sequential data without recurrent layers.                                   |


# Key Machine Learning & Data Science Concepts (ML Learner Level)

| Concept                     | Simple Definition                                                               |
| --------------------------- | ------------------------------------------------------------------------------- |
| **Overfitting**             | The model memorizes training data too well but fails on new data.               |
| **Underfitting**            | The model is too simple and misses patterns in data.                            |
| **Training Data**           | The data used to teach the model.                                               |
| **Testing Data**            | New data used to see how good the model is.                                     |
| **Cross-Validation**        | A method to test the model on different splits of the data to be more accurate. |
| **Bias**                    | Error caused by overly simple assumptions.                                      |
| **Variance**                | Error caused by too much complexity (model is too sensitive).                   |
| **Hyperparameters**         | Settings you choose before training (like how many trees in a forest).          |
| **Features**                | Inputs to your model (columns like age, income, etc.).                          |
| **Target/Label**            | What you're trying to predict (like price or category).                         |
| **Accuracy**                | % of predictions the model got right.                                           |
| **Precision**               | Of the things the model said were positive, how many actually were?             |
| **Recall**                  | Of all the actual positives, how many did the model catch?                      |
| **F1 Score**                | Balance between precision and recall.                                           |
| **ROC Curve / AUC**         | Measures how well the model separates different classes.                        |
| **Confusion Matrix**        | A table showing true vs predicted values.                                       |
| **Feature Engineering**     | Creating new input features to help the model.                                  |
| **Regularization**          | Keeps the model from overfitting by reducing large weights.                     |
| **Imputation**              | Filling in missing data.                                                        |
| **One-Hot Encoding**        | Turns categories into numbers so models can understand them.                    |
| **Normalization / Scaling** | Reshapes numbers to be in the same range.                                       |
| **Pipeline**                | A step-by-step process to prepare data and train the model.                     |
| **Model Evaluation**        | Checking how well your model works.                                             |
| **Deployment**              | Putting your trained model into a real-world app or system.                     |
